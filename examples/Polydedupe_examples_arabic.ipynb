```python
from PolyDeDupe import deduplicate_dataset, display_dataset_entries
from datasets import load_dataset


def build_dataset(data):
    system = next((convo["value"] for convo in data["conversations"] if convo["from"] == "system"), "")
    instruction = next((convo["value"] for convo in data["conversations"] if convo["from"] == "human"), "")
    assistant_or_agent_convo = next(
        (convo["value"] for convo in data["conversations"] if convo["from"] in ["gpt", "assistant"]), ""
    )
    data["text"] = f"""System: {system} Human: {instruction} Assitant: {assistant_or_agent_convo}"""
    return data


dataset = load_dataset("FreedomIntelligence/alpaca-gpt4-arabic", split="train")
newdataset = dataset.map(build_dataset, num_proc=16, remove_columns=dataset.column_names)
ds_dedup, duplicate_clusters = deduplicate_dataset(newdataset, jaccard_threshold=0.90)
```

```python
display_dataset_entries(newdataset, duplicate_clusters)
```

```python
dataset = load_dataset("Ali-C137/Hindawi-Books-dataset", split="train[:1000]")

newdataset = dataset.rename_column("ChapterText", "text")
newdataset = newdataset.filter(lambda example: example["text"] is not None)

newdataset = newdataset.remove_columns(
    ["BookLink", "BookName", "AuthorName", "AboutBook", "ChapterLink", "ChapterName", "AboutAuthor"]
)
ds_dedup, duplicate_clusters = deduplicate_dataset(newdataset, jaccard_threshold=0.90)
```
